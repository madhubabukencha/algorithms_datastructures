In computer science, a <b>Linear Search</b> or <b>Sequencial Search</b> is a method for finding an element with in a list. It sequencially checks each element of the list until a match found or the whole list has been searched. If the algorithm reaches the end of the list, the search terminate unsuccessfully

```python
def linear_search(my_list, element_to_find):
    """
    Linear Search:
    If value found in the list then it will return print the index
    else it if will print "Element not found" message
    """
    len_of_list = len(my_list)
    for i in range(len_of_list):
        if my_list[i] == element_to_find:
            print(f"The index of element {element_to_find} is {i}")
            break
    else:
        print("Element not found")
```

```python
linear_search([3, 4, 5, 6, 1], 1)
```

```python
linear_search([3, 4, 5, 6, 1], 8)
```
