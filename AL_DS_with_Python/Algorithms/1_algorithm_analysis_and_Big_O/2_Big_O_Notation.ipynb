<b>What is Big O ?</b> <br>
Big O is a formal notation that describes the behaviour of a function when the argument tends towards the maximum input. <br>
Big O takes the upper bound, which means the worst case results in the worst execution of algorithm. If we take shopping <br>
as example, worst case is infinite number of items.<br><br>

Assume input of 100 million or infinite etc.., = <b>n</b> <br><br>

<b>Big O notation order of growth table:</b><br>
We will use these Big O notation functions to calculate algorithem efficiency
 <br>

<b>Time taken by each Big O notation</b>:
Constant < Logarithm < Linear < Polynamial < Exponential <br>

<p>If we take our shopping items list, assume that in worst case our algorithm printing all items in the list sequentially. Since <br> there are  n items in the shopping items list it takes O(n) to complete the algorithm</p>
    

<b>Other Asymptotic(time measuring) notation are:</b>
 <br>
<b>Informally</b>: Big Omega(best case), Big Theta(Average case), Big O(Worst Case)

<font size=5 color="color:#290001;">Big-O visualization:</font>

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("bmh")

n = np.linspace(1, 10, 1000)
labels = ['Constant','Logarithmic','Linear','Log Linear','Quadratic','Cubic','Exponential']
big_o = [np.ones(n.shape),np.log(n),n,n*np.log(n),n**2,n**3,2**n]

# Plot setup
plt.figure(figsize=(12,10))
plt.ylim(1,50)

for i in range(len(big_o)):
    plt.plot(n,big_o[i],label = labels[i])


plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('n')
```
