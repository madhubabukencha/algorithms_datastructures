<strong>Definition:</strong>The element which we entered first, will be removed at the last

```python
from IPython.display import Image
Image(filename='images/stack.gif')
```

```python
container = []
index = -1

def insert(element):
    """
    To insert elements in the list
    """
    global index
    index += 1
    container.append(element)


def pop():
    """
    To remove a element from the list
    """
    global index
    if index == -1:
        print("Container is empty")
    else:
        del container[index]
        index -= 1
```

```python
insert(10)
insert(11)
insert(12)
insert(14)
container
```

```python
pop()
container
```

<font size=5 color="color:#290001;">Time Complexity:</font>

The insertion and deletion happens only at the top and once at a time. So, time complexity is O(1)
